```python
# define number of players
n_players = 8
n_rounds = 5

# set players per match
ppm = 4
```

```python
# get list of letters
import string
all_players = list(string.ascii_uppercase)
players = all_players[:n_players]
print(players)
```

```python
import random

def player_check(occurances, round_players, player):
    """
    given a list and a character, return the number of occurances in the list
    """
    nocc = occurances.count(player)
    
    if (nocc < 4) & (player not in round_players):
        return True
    else:
        return False
    
def team_check(teams, players):
    """
    given a list of teams and two players, check they have never played together
    """
    for team in teams:
        if team == players:
            return False
    return True


# big list to count occurences of players and teams drawn
occurances = []
teams = []

# get the number of matches in each round
matches_per_round = int(n_players / ppm)

# go through rounds
for m in range(n_rounds):
    
    print('\n')
    print('Round {} ================'.format(m+1))
    print('\n')
    
    round_players = []
    
    # for each round make teams
    for r in range(matches_per_round):

        match_team1 = []
        match_team2 = []

        while True:
            while True:
                player_1 = players[random.randint(1, n_players)-1] 
                if player_check(occurances, round_players, player_1):
                    break

            while True:
                player_2 = players[random.randint(1, n_players)-1] 
                if player_check(occurances, round_players, player_2):
                    break

            if player_2 != player_1:
                if team_check(teams, [player_1, player_2]) | team_check(teams, [player_2, player_1]):
                    break

        match_team1.append(player_1)
        match_team1.append(player_2)
        round_players.append(player_1)
        round_players.append(player_2)


        while True:
            while True:
                player_3 = players[random.randint(1, n_players)-1] 
                if player_check(occurances, round_players, player_3):
                    break

            while True:
                player_4 = players[random.randint(1, n_players)-1] 
                if player_check(occurances, round_players, player_4):
                    break

            if player_3 != player_4:
                if team_check(teams, [player_3, player_4]) | team_check(teams, [player_4, player_3]):
                    break

        match_team2.append(player_3)
        match_team2.append(player_4)
        round_players.append(player_3)
        round_players.append(player_4)
            
        teams.append(match_team1)
        teams.append(match_team2)
                
        print('Match {}:'.format(r+1))
        print('{} & {}  vs.  {} & {}'.format(match_team1[0],match_team1[1],match_team2[0],match_team2[1]))
        print('\n')
        
    # add each player to the total list
    occurances.append(player_1)
    occurances.append(player_2)
    occurances.append(player_3)
    occurances.append(player_4)
```
